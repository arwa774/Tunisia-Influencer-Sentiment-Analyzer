```python
import pandas as pd

# ── 1. Map sentiment labels to numeric values ──────────────────────────────
sentiment_map = {"Positive": 1, "Neutral": 0, "Negative": -1}
df["sentiment_value"] = df["predicted_sentiment"].map(sentiment_map)
username_column="username"
# ── 2. Per-user sentiment counts & percentages ────────────────────────────
total        = df.groupby(username_column).size().rename("total_comments")
pos_count    = df[df["predicted_sentiment"] == "Positive"].groupby(username_column).size().rename("positive_count")
neg_count    = df[df["predicted_sentiment"] == "Negative"].groupby(username_column).size().rename("negative_count")
neu_count    = df[df["predicted_sentiment"] == "Neutral" ].groupby(username_column).size().rename("neutral_count")

user_df = pd.concat([total, pos_count, neg_count, neu_count], axis=1).fillna(0)

user_df["positive_pct"] = (user_df["positive_count"] / user_df["total_comments"] * 100).round(2)
user_df["negative_pct"] = (user_df["negative_count"] / user_df["total_comments"] * 100).round(2)
user_df["neutral_pct"]  = (user_df["neutral_count"]  / user_df["total_comments"] * 100).round(2)

# ── 3. Confidence-Weighted Sentiment Score ────────────────────────────────
#
#   score_i  = confidence_i × sentiment_value_i      (per comment)
#
#   where  sentiment_value ∈ { +1 (Positive), 0 (Neutral), −1 (Negative) }
#
#   User Score = ( Σ score_i / n ) × 100
#
#   • Range  : −100  →  +100
#   • +100   : every comment is Positive with 100 % confidence
#   • −100   : every comment is Negative with 100 % confidence
#   •   0    : perfectly balanced or all Neutral
#   • Confidence acts as a reliability weight — a low-confidence
#     prediction contributes less to the final score than a
#     high-confidence one.
# ─────────────────────────────────────────────────────────────────────────
weighted_score = (
    df.groupby(username_column)
      .apply(lambda g: (g["confidence"] * g["sentiment_value"]).sum() / len(g) * 100)
      .rename("sentiment_score")
      .round(2)
)

user_df = user_df.join(weighted_score)

# ── 4. Clean up & export ──────────────────────────────────────────────────
result = user_df[
    ["total_comments",
     "positive_count", "positive_pct",
     "negative_count", "negative_pct",
     "neutral_count",  "neutral_pct",
     "sentiment_score"]
].reset_index()

result.to_csv("user_sentiment_summary_youtube.csv", index=False)
print(result)
```

```python
import pandas as pd

# ── 1. Load both CSVs ──────────────────────────────────────────────────────
yt_df  = pd.read_csv("user_sentiment_summary_youtube.csv")
tt_df  = pd.read_csv("user_sentiment_summary_tiktok (2).csv")

# ── 2. Rename columns with platform prefix (keep user_name as merge key) ──
yt_df  = yt_df.rename(columns={col: f"youtube_{col}" for col in yt_df.columns if col != "user_name"})
tt_df  = tt_df.rename(columns={col: f"tiktok_{col}"  for col in tt_df.columns if col != "user_name"})

# ── 3. Outer join on user_name ─────────────────────────────────────────────
#   - users on BOTH platforms  → all columns filled
#   - users on ONE platform    → missing platform columns → filled with 0
global_df = pd.merge(yt_df, tt_df, on="user_name", how="outer").fillna(0)

# ── 4. Global score ────────────────────────────────────────────────────────
#
#   Case A  – user exists on BOTH platforms:
#             global_score = (youtube_score + tiktok_score) / 2
#
#   Case B  – user exists on ONE platform only:
#             global_score = the available platform score
#             (the other score is 0, so average would be wrong → use the real one)
#
#   Detection: if total_comments for a platform is 0, the user wasn't there.
# ──────────────────────────────────────────────────────────────────────────
def compute_global_score(row):
    on_youtube = row["youtube_total_comments"] > 0
    on_tiktok  = row["tiktok_total_comments"]  > 0

    if on_youtube and on_tiktok:
        return round((row["youtube_sentiment_score"] + row["tiktok_sentiment_score"]) / 2, 2)
    elif on_youtube:
        return round(row["youtube_sentiment_score"], 2)
    else:
        return round(row["tiktok_sentiment_score"], 2)

global_df["global_score"] = global_df.apply(compute_global_score, axis=1)

# ── 5. Reorder columns neatly ─────────────────────────────────────────────
youtube_cols = [c for c in global_df.columns if c.startswith("youtube_")]
tiktok_cols  = [c for c in global_df.columns if c.startswith("tiktok_")]

global_df = global_df[["user_name"] + youtube_cols + tiktok_cols + ["global_score"]]

# ── 6. Save ───────────────────────────────────────────────────────────────
global_df.to_csv("user_sentiment_global.csv", index=False)
print(f"✅ Saved → user_sentiment_global.csv  ({len(global_df)} users)")
print(global_df.head())
```
